# Comparing Results

Which API provided you with more complete data? Provide an explanation. 

Get the top 10 restaurants according to their rating

```python
import requests

def get_top_restaurants(city, radius, api_key):
    url = "https://api.yelp.com/v3/businesses/search"
    headers = {
        "Authorization": f"Bearer {api_key}"
    }
    params = {
        "location": city,
        "radius": radius,
        "categories": "restaurants", # Filtering to only resturants
        "sort_by": "rating",        # soting by the ratings
        "limit": 10                 # limiting to 10
    }

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    if response.status_code == 200:
        businesses = data["businesses"]
        for business in businesses:
            name = business["name"]
            rating = business["rating"]
            address = ", ".join(business["location"]["display_address"])
            print(f"Name: {name}, Rating: {rating}, Address: {address}")
    else:
        print(f"Error occurred while fetching top restaurants. Status Code: {response.status_code}")

# Replace "YOUR_API_KEY" with your actual Yelp API key
api_key = "dlLOQLSDztStbXVlLma_v79dN5EdTJTz5jgzOQ5-KKIC2g-T5Ja6pAGLrEMzbM--YVOTdoj2y-1lsnupn_Ua4joo0NCkx1K_qczFz97l7joDinygxXYjO-IbYPxwZXYx"
get_top_restaurants("Toronto", 1000, api_key)
```
