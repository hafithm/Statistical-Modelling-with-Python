```python
import requests
import os
import pandas as pd


def get_bike_stations(city, categories):
    url = "https://api.foursquare.com/v3/places/search"

    # Retrieve Foursquare API key from environment variable
    fsq_api_key = os.environ['FOURSQUARE_API_KEY']

    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {fsq_api_key}"
    }
    params = {
        "ll": "43.6532,-79.3832",
        "radius": 1000,
        "limit": 50,
        "categories": categories
    }

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    extracted_data = []
    for item in data.get('results', []):
        extracted_data.append({
            'name': item.get('name'),
            'location': item.get('location', {}).get('address'),
            'rating': item.get('rating'),
            'popularity': item.get('popularity'),
            'price': item.get('price'),
            'station_id': item.get('station_id'),
            'fsq_id': item.get('fsq_id'),
            'distance': item.get('distance'),
            'latitude': "43.6532",
            'longitude': "-79.3832"
        })

    return extracted_data

city = "Toronto"
requested_data_restaurants = get_bike_stations(city, 13065)
requested_data_coffee_shops = get_bike_stations(city, 13032)

df_restaurants = pd.DataFrame(requested_data_restaurants)
df_coffee_shops = pd.DataFrame(requested_data_coffee_shops)


print("All restaurants and Coffee Cafes")

# Concatenate the dataframes

foursquare_data = pd.concat([df_restaurants, df_coffee_shops])

foursquare_data
```

```python
import requests
import pandas as pd

yelp_api_key = os.environ['YELP_API_KEY']
url = 'https://api.yelp.com/v3/businesses/search'
headers = {
    'Authorization': f'Bearer {yelp_api_key}',
    'accept': 'application/json',
}

# Get restaurants data
params_restaurants = {
    'location': 'toronto',
    'latitude': 43.66,
    'longitude': -79.31,
    'term': 'restaurant', 
    'radius': 1000,
    'categories': 'restaurants, cafe',
    'sort_by': 'rating',
    'limit': 50,
}

response_restaurants = requests.get(url, params=params_restaurants, headers=headers)
data_restaurants = response_restaurants.json()

# Get coffee shops data
params_coffee_shops = {
    'location': 'toronto',
    'latitude': 43.66,
    'longitude': -79.31,
    'term': 'coffee', 
    'radius': 1000,
    'categories': 'restaurants, cafe',
    'sort_by': 'rating',
    'limit': 50,
}

response_coffee_shops = requests.get(url, params=params_coffee_shops, headers=headers)
data_coffee_shops = response_coffee_shops.json()

# Create dataframes
df_restaurants2 = pd.json_normalize(data_restaurants['businesses'])
df_coffee_shops2 = pd.json_normalize(data_coffee_shops['businesses'])

# Rename columns
df_restaurants2.rename(columns={'coordinates.latitude': 'latitude', 'coordinates.longitude': 'longitude'}, inplace=True)
df_coffee_shops2.rename(columns={'coordinates.latitude': 'latitude', 'coordinates.longitude': 'longitude'}, inplace=True)

# Drop columns
df_restaurants2.drop(['location.address2', 'location.address3'], axis=1, inplace=True)
df_coffee_shops2.drop(['location.address2', 'location.address3'], axis=1, inplace=True)

# Concatenate the dataframes
yelp_df = pd.concat([df_restaurants2, df_coffee_shops2])

print("All resturants and coffee cafes")
yelp_df
```

```python
import requests
import pandas as pd

yelp_api_key = os.environ['YELP_API_KEY']
url = 'https://api.yelp.com/v3/businesses/search'
params = {
    'location': 'toronto',
    'latitude': 43.66,
    'longitude': -79.31,
    'term': 'restaurant',
    'radius': 1000,
    'sort_by': 'rating',
    'limit': 10,  # Update limit to 10 #indicating that we want to retrieve only the top 10 restaurants
}
headers = {
    'Authorization': f'Bearer {yelp_api_key}',
    'accept': 'application/json',
}

response = requests.get(url, params=params, headers=headers)
data = response.json()

yelprestaurant_df = pd.json_normalize(data['businesses'])

# Sort the dataframe by rating in descending order
yelprestaurant_df = yelprestaurant_df.sort_values('rating', ascending=False)

yelprestaurant_df[['name', 'rating']]
```
