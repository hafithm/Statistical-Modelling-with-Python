```python
import sqlite3
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/Users/hm4th/downloads/merged_df2.csv'
merged_df2 = pd.read_csv(file_path)

# Rename 'Location' column to 'address'
merged_df2.rename(columns={'Location': 'address'}, inplace=True, errors='ignore')

# Connect to a new SQLite database in your directory
new_conn = sqlite3.connect('/Users/hm4th/Downloads/mydata.db')

# Save the DataFrame to the SQLite database
with new_conn:
    merged_df2.to_sql('merged_df2', new_conn, if_exists='replace', index=False)

# Reconnect to the newly created SQLite database
new_conn = sqlite3.connect('/Users/hm4th/Downloads/mydata.db')

# Execute a SELECT query to fetch the data from the newly created table
query = 'SELECT * FROM merged_df2'  # Assuming 'merged_df2' is the table name
df_from_db = pd.read_sql_query(query, new_conn)

# Close the connection to the new database
new_conn.close()

# Print the DataFrame fetched from the database
print(df_from_db)


```

```python
import sqlite3
import pandas as pd

# Connect to the existing SQLite database
existing_conn = sqlite3.connect('/Users/hm4th/Downloads/mydata.db')

# Create a cursor object for the existing database
existing_cursor = existing_conn.cursor()

# Execute a SELECT query to fetch the data from the existing table
existing_cursor.execute('SELECT rating, free_bikes, "bike station", rating FROM merged_df2')

# Fetch all the rows from the result set
data = existing_cursor.fetchall()

# Close the cursor and connection to the existing database
existing_cursor.close()
existing_conn.close()

# Create a dataframe from the fetched data
df = pd.DataFrame(data, columns=[ 'distance', 'free_bikes', 'rating' , 'bike station'])

# Calculate correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print(correlation_matrix)

```
